```python
# Desired order of room_type values
room_type_order = ['Private room', 'Entire home/apt', 'Shared room', 'Hotel room']

# Convert room_type to a categorical type with specified order
df['room_type'] = pd.Categorical(df['room_type'], categories=room_type_order, ordered=True)

# Rearrange the data by neighbourhood_group, neighbourhood, and room_type
df_rearranged = df.sort_values(by=['neighbourhood_group', 'neighbourhood', 'room_type']).reset_index(drop=True)
df['neighbourhood_group'] = df['neighbourhood_group'].replace('brookln', 'Brooklyn')

# Delete rows where availability_365 is less than 60
df = df[df['availability_365'] >= 60]

# Fill missing values in the reviews_per_month column with 0
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)

# Round values in the reviews_per_month column to a single whole number
df['reviews_per_month'] = df['reviews_per_month'].round(0).astype(int)

# Convert the host_name column to a string
df.loc[:,'host_name'] = df['host_name'].astype(str)

# Filter out rows where host_name starts with a number or a special character
df = df[df['host_name'].apply(lambda x: str(x)[0].isalpha())]
df = df[df['host_name'].str.match(r'^[a-zA-Z].*')]
df['minimum_nights'] = df['minimum_nights'].fillna(1.0)
```

```python
# Checking for outliers
# Tukey's method for identifying outliers.
# Threshold value of 1.5 times the IQR
for column in df[['price', 'number_of_reviews', 'minimum_nights']]:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    threshold = 1.5
    outliers = df[(df[column] < Q1 - threshold * IQR) | (df[column] > Q3 + threshold * IQR)]
    print(column)
    print(outliers.shape[0])
```

```python
# Investigating outliers in 'number_of_reviews
Q1 = df['number_of_reviews'].quantile(0.25)
Q3 = df['number_of_reviews'].quantile(0.75)
IQR = Q3 - Q1
threshold = 1.5
outliers = df[(df['number_of_reviews'] < Q1 - threshold * IQR) | (df['number_of_reviews'] > Q3 + threshold * IQR)]
outliers.sort_values(by='number_of_reviews', ascending= False)
```

```python
# Fill missing values in the 'price' column
# Replace any missing (NaN) values in the 'price' column using the forward fill method ('ffill'),
# Which propagates the last observed non-null value forward until another non-null value is encountered.
df['price'] = df['price'].fillna(method='ffill')
```

```python
#cheking for inconsistencies in quantitive data
# Checks if values in the latitude column fall between -90 and 90 degrees.
# Latitude values outside this range are invalid as they represent locations beyond the Earth's poles.
# Similarly, this part checks if the values in the longitude column fall between -180 and 180 degrees.
df = df[(df['latitude'].between(-90,90) & df['longitude'].between(-180,180))]
```

```python
# Function to categorize time periods.
def categorize_time(time):
  # Extract the hour from the time string
    hour = int(time.split(':')[0])
    # Categorize the time into 'Early Morning', 'Morning', 'Afternoon', or 'Evening/Night'
    if 0 <= hour < 6:
        return 'Early Morning'
    elif 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    else:
        return 'Evening/Night'
df_sh['time_period'] = df_sh['OCCUR_TIME'].apply(categorize_time)
```

### Average Reviews per Month by Date and Room Type - Join on Neighborhood Table

```python
def avg_rv_date(conn):
    """ Query the first 10 rows from the Neighborhoods table """
    try:
        cur = conn.cursor()
        query = """
       SELECT
  l.date,
  l.room_type,
  AVG(l.reviews_per_month) AS avg_reviews_per_month
FROM listings l
INNER JOIN Neighborhoods n ON l.incident_key = n.incident_key
GROUP BY l.date, l.room_type
ORDER BY l.date, l.room_type;
        """
        cur.execute(query)
        rows = cur.fetchmany(10)
        for row in rows:
            print(row)
        cur.close()
    except (Exception, psycopg2.DatabaseError) as error:
        print(error)
conn = connect_to_db(params)

if conn is not None:
    avg_rv_date(conn)
    conn.close()
else:
    print("Failed to connect to the database")
```

### Average Price per Room Type (without Neighborhood Join):

```python
def avg_price_pr(conn):
    """ Query the first 10 rows from the Neighborhoods table """
    try:
        cur = conn.cursor()
        query = """
   SELECT l.room_type, AVG(l.price_y) AS avg_price
FROM listings l
GROUP BY l.room_type
ORDER BY avg_price;
        """
        cur.execute(query)
        rows = cur.fetchmany(10)
        for row in rows:
            print(row)
        cur.close()
    except (Exception, psycopg2.DatabaseError) as error:
        print(error)
conn = connect_to_db(params)

if conn is not None:
    avg_price_pr(conn)
    conn.close()
else:
    print("Failed to connect to the database")
```

##Monthly Listing Count:

```python
def moving_out(conn):
    """ Query the first 10 rows from the Neighborhoods table """
    try:
        cur = conn.cursor()
        query = """
SELECT DATE_TRUNC('month', l.date) AS listing_month, COUNT(*) AS listing_count
FROM listings l
GROUP BY DATE_TRUNC('month', l.date)
ORDER BY listing_month;
        """
        cur.execute(query)
        rows = cur.fetchmany(10)
        for row in rows:
            print(row)
        cur.close()
    except (Exception, psycopg2.DatabaseError) as error:
        print(error)


conn = connect_to_db(params)
if conn is not None:
    moving_out(conn)
    conn.close()
else:
    print("Failed to connect to the database")
```
